# Transforming

```python
#train test split - import 
from sklearn.model_selection import train_test_split as tts

x, x_test, y, y_test = tts(x_credit, credit_target, test_size=0.20, train_size=0.80, random_state=33) 
x_train, x_val, y_train, y_val = tts(x, y, test_size=0.25, train_size=0.75, random_state=33)
```

```python
print("Train size:", x_train.shape)
print("Test size:", x_test.shape)
print("Validation size:", x_val.shape)
```

```python
from sklearn.preprocessing import LabelBinarizer, StandardScaler, LabelEncoder

SS = StandardScaler()
X_train = SS.fit_transform(x_train)
X_test  = SS.fit_transform(x_test)
X_val   = SS.fit_transform(x_val)

LB = LabelBinarizer()

Y_train = LB.fit_transform(y_train)
Y_val   = LB.fit_transform(y_val)
Y_test  = LB.fit_transform(y_test)
```

# The model

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, LeakyReLU
from keras.optimizers import Adam, SGD
from sklearn.metrics import accuracy_score
from tensorflow.keras.callbacks import EarlyStopping
from keras.layers.normalization import BatchNormalization

baseline = Sequential()
#INPUT LAYER
baseline.add(Dense(96, input_dim=X_train.shape[1], activation='tanh'))


#HIDDEN LAYERS
baseline.add(Dense(64)) #, activation="relu"))
baseline.add(LeakyReLU(alpha=0.1))
baseline.add(Dropout(0.3))
baseline.add(BatchNormalization())

baseline.add(Dense(64)) #, activation="relu"))
baseline.add(LeakyReLU(alpha=0.1))
baseline.add(Dropout(0.3))
baseline.add(BatchNormalization())

baseline.add(Dense(32)) #, activation="relu"))
baseline.add(LeakyReLU(alpha=0.1))
baseline.add(Dropout(0.3))
baseline.add(BatchNormalization())

"""baseline.add(Dense(32))
baseline.add(LeakyReLU(alpha=0.1))
baseline.add(Dropout(0.1))
baseline.add(BatchNormalization())
"""

#OUTPUT LAYER
baseline.add(Dense(1, activation='sigmoid')) 

#run the model
optimizer = Adam(lr=0.001) 
baseline.compile(optimizer=optimizer, loss="binary_crossentropy", metrics="accuracy")
history = baseline.fit(X_train, Y_train, epochs=30, batch_size=32, verbose=1, validation_data=(X_test,Y_test), 
           callbacks=[EarlyStopping(monitor='val_loss', patience=1)])

#loss: 0.5361 - accuracy: 0.7450 - val_loss: 0.4778 - val_accuracy: 0.7900
```

```python
import matplotlib.pyplot as plt 

#accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

#loss 
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

# Balanced Residuals

## baseline

```python
import numpy as np

y_pred_val   = (baseline.predict(X_val) > 0.5).astype("int32")
y_hats_val   = pd.DataFrame(data = y_pred_val, columns=['y_pred_val'], index=x_val.index.copy())
```

```python
regular_out = pd.merge(x_credit, y_hats_val, how='left', left_index=True, right_index=True )

regular_out["y_val"] = y_val

display(regular_out)
```

```python
cleanup_nums1 = {"y_pred_val": {0.0:1, 1.0:2}}

regular_out = regular_out.replace(cleanup_nums1)
display(regular_out)
```

```python
age1 = regular_out[regular_out["Age"] == 1]
age2 = regular_out[regular_out["Age"] == 2]

age_young = pd.concat([age1, age2])
display(age_young)

age3 = regular_out[regular_out["Age"] == 3]
age4 = regular_out[regular_out["Age"] == 4]
age5 = regular_out[regular_out["Age"] == 5]

age_old = pd.concat([age3, age4, age5])
display(age_old)
```

```python
age_old_ypredval = age_old["y_pred_val"]
age_old_ytrueval = age_old["y_val"]

age_young_ypredval = age_young["y_pred_val"]
age_young_ytrueval = age_young["y_val"]

age_old_ypredval = age_old_ypredval.dropna()
age_old_ytrueval = age_old_ytrueval.dropna()

age_young_ypredval = age_young_ypredval.dropna()
age_young_ytrueval = age_young_ytrueval.dropna()

display(age_old_ypredval)
display(age_old_ytrueval)
display(age_young_ypredval)
display(age_young_ytrueval)
```

```python
# TRUE - PRED
val_dif_old = (age_old_ytrueval - age_old_ypredval).sum()
print(val_dif_old)

residuals_old = val_dif_old / 88 
print(residuals_old)

# TRUE - PRED
val_dif_young = (age_young_ytrueval - age_young_ypredval).sum()
print(val_dif_young)

residuals_young = val_dif_young / 112 
print(residuals_young)
```

```python
discrimination_balanced_residuals_regular = residuals_young - residuals_old

print(discrimination_balanced_residuals_regular)
```
