```python
import pandas as pd
pd.set_option('display.max_columns', None)

credit = pd.read_csv("german.csv")

display(credit)

#Target: 1=good, 2=bad
```

```python
cleanup_nums = {"StatusExistingCheckingAccount": {"A11":1, "A12":2, "A13":3, "A14":4},
                "CreditHistory": {"A30":1, "A31":2, "A32":3, "A33":4, "A34":5},
                "Purpose": {"A40":1, "A41":2, "A42":3, "A43":4, "A44":5, "A45":6, "A46":7, "A47":8, "A48":9, "A49":10, "A410":11},
                "SavingsAccountBonds": {"A61":1, "A62":2, "A63":3, "A64":4, "A65":5},
                "PresentEmploymentSince": {"A71":1, "A72":2, "A73":3, "A74":4, "A75":5},
                "PersonalStatusAndSex": {"A91":1, "A93":1, "A94":1, "A92":2, "A95":2},
                "OtherDebtorsGuarantors": {"A101":1, "A102":2, "A103":3},
                "Property": {"A121":1, "A122":2, "A123":3, "A124":4},
                "OtherInstallmentPlans": {"A141":1, "A142":2, "A143":3},
                "Housing": {"A151":1, "A152":2, "A153":3},
                "Job": {"A171":1, "A172":2, "A173":3, "A174":4},
                "Telephone": {"A191":1, "A192":2},
                "ForeignWorker": {"A201":1, "A202":2}}

x_credit = credit.replace(cleanup_nums)
display(x_credit)
```

```python
x_credit["Age"] = pd.cut(x_credit.Age,bins=[0,24,34,44,54, 100],labels=[1, 2, 3, 4, 5])
```

```python
x_credit['Age'].value_counts()
```

```python
cleanup_nums2 = {"Age": {1:1, 2:1, 3:2, 4:2, 5:2}}

x_credit = x_credit.replace(cleanup_nums2)
display(x_credit)
```

```python
x_credit['Age'].value_counts()
```

# Fair Approach

```python
x_credit1 = x_credit[x_credit["Age"] == 1]
x_credit2 = x_credit[x_credit["Age"] == 2]
display(x_credit1)
display(x_credit2)
```

```python
x_credit1_sample = x_credit1.sample(n=452)
display(x_credit1_sample)
```

```python
x_credit_new = pd.concat([x_credit1_sample, x_credit2])
display(x_credit_new)
```

```python
pd.crosstab(x_credit_new.CreditApproved, x_credit_new.Age)
```

```python
compas_target = x_credit_new['CreditApproved'] 
display(compas_target)
```

```python
x_credit_new2 = x_credit_new.drop(columns=["CreditApproved"])
display(x_credit_new2)
```

# Transforming

```python
#train test split - import 
from sklearn.model_selection import train_test_split as tts

x, x_test, y, y_test = tts(x_credit_new2, compas_target, test_size=0.20, train_size=0.80, random_state=33)
x_train, x_val, y_train, y_val = tts(x, y, test_size=0.25, train_size=0.75, random_state=33)

print("Train size:", x_train.shape)
print("Test size:", x_test.shape)
print("Validation size:", x_val.shape)
```

```python
pd.crosstab(y_train, x_train.Age)
```

```python
x_train_score = x_train.copy()
display(x_train_score)
```

```python
x_train_score["y_train"] = y_train 
display(x_train_score)
```

```python
x_trainA1   = x_train_score[x_train_score["Age"] == 1]
x_trainA2   = x_train_score[x_train_score["Age"] == 2]

display(x_trainA1)
display(x_trainA2)
```

```python
A1_bad = x_trainA1[x_trainA1["y_train"] == 2]
A2_good = x_trainA2[x_trainA2["y_train"] == 1]

display(A1_bad)
display(A2_good)
```

```python
A1_bad_sample      = A1_bad.sample(n=12)
A2_good_sample     = A2_good.sample(n=12)
```

```python
display(A1_bad_sample)
```

```python
cleanup_nums_bad = {"y_train": {2:1}}

A1_bad_sample = A1_bad_sample.replace(cleanup_nums_bad)
display(A1_bad_sample)
```

```python
display(A2_good_sample)
```

```python
cleanup_nums_good = {"y_train": {1:2}}

A2_good_sample = A2_good_sample.replace(cleanup_nums_good)
display(A2_good_sample)
```

```python
merged = pd.concat([A1_bad_sample, A2_good_sample])

display(merged)
```

```python
A1BS  = A1_bad_sample.index
A2BS  = A2_good_sample.index
```

```python
x_train_score.drop(A1BS, inplace=True)
x_train_score.drop(A2BS, inplace=True)
```

```python
display(x_train_score)
```

```python
fair_train_score = pd.concat([x_train_score, merged])
display(fair_train_score)
```

```python
pd.crosstab(fair_train_score.y_train, fair_train_score.Age)
```

```python
fair_y_train = fair_train_score["y_train"]
display(fair_y_train)
```

```python
fair_train = fair_train_score.drop(columns=["y_train"])
display(fair_train)
```

```python
from sklearn.preprocessing import LabelBinarizer, StandardScaler, LabelEncoder

SS = StandardScaler()
X_train_fair = SS.fit_transform(fair_train)
X_test       = SS.fit_transform(x_test)
X_val        = SS.fit_transform(x_val)

LB = LabelBinarizer()

Y_train_fair = LB.fit_transform(fair_y_train)
Y_val        = LB.fit_transform(y_val)
Y_test       = LB.fit_transform(y_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, LeakyReLU
from keras.optimizers import Adam, SGD
from sklearn.metrics import accuracy_score
from tensorflow.keras.callbacks import EarlyStopping
from keras.layers.normalization import BatchNormalization

fair = Sequential()
#INPUT LAYER
fair.add(Dense(96, input_dim=X_train_fair.shape[1], activation='tanh'))


#HIDDEN LAYERS
fair.add(Dense(64)) #, activation="relu"))
fair.add(LeakyReLU(alpha=0.1))
fair.add(Dropout(0.3))
fair.add(BatchNormalization())

fair.add(Dense(64)) #, activation="relu"))
fair.add(LeakyReLU(alpha=0.1))
fair.add(Dropout(0.3))
fair.add(BatchNormalization())

fair.add(Dense(32)) #, activation="relu"))
fair.add(LeakyReLU(alpha=0.1))
fair.add(Dropout(0.3))
fair.add(BatchNormalization())


#OUTPUT LAYER
fair.add(Dense(1, activation='sigmoid')) 

#run the model
optimizer = Adam(lr=0.001) 
fair.compile(optimizer=optimizer, loss="binary_crossentropy", metrics="accuracy")
history = fair.fit(X_train_fair, Y_train_fair, epochs=30, batch_size=32, verbose=1, validation_data=(X_test,Y_test), 
           callbacks=[EarlyStopping(monitor='val_loss', patience=1)])
```

```python
import matplotlib.pyplot as plt 

#accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

#loss 
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

# Balanced Residuals

```python
y_pred_val   = (fair.predict(X_val) > 0.5).astype("int32")
y_hats_val   = pd.DataFrame(data = y_pred_val, columns=['y_pred_val'], index=x_val.index.copy())
```

```python
regular_out = pd.merge(x_credit_new2, y_hats_val, how='left', left_index=True, right_index=True )

regular_out["y_val"] = y_val

display(regular_out)
```

```python
cleanup_nums = {"y_pred_val": {0.0:1, 1.0:2}}

regular_out = regular_out.replace(cleanup_nums)
display(regular_out)
```

```python
age1 = regular_out[regular_out["Age"] == 1]
age2 = regular_out[regular_out["Age"] == 2]

display(age1)
display(age2)
```

```python
age2_ypredval = age2["y_pred_val"]
age2_ytrueval = age2["y_val"]

age1_ypredval = age1["y_pred_val"]
age1_ytrueval = age1["y_val"]

age2_ypredval = age2_ypredval.dropna()
age2_ytrueval = age2_ytrueval.dropna()

age1_ypredval = age1_ypredval.dropna()
age1_ytrueval = age1_ytrueval.dropna()

display(age2_ypredval)
display(age2_ytrueval)
display(age1_ypredval)
display(age1_ytrueval)
```

```python
# TRUE - PRED
val_dif_old = (age2_ytrueval - age2_ypredval).sum()
print(val_dif_old)

residuals_old = val_dif_old / 92 
print(residuals_old)

# TRUE - PRED
val_dif_young = (age1_ytrueval - age1_ypredval).sum()
print(val_dif_young)

residuals_young = val_dif_young / 89 
print(residuals_young)
```

```python
discrimination_balanced_residuals_protected = residuals_young - residuals_old

print(discrimination_balanced_residuals_protected)
```
