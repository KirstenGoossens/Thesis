```python
from sklearn.preprocessing import LabelBinarizer, StandardScaler, LabelEncoder, MinMaxScaler

MMS = MinMaxScaler()
X_train_fair = MMS.fit_transform(fair_train)
X_test       = MMS.fit_transform(x_test)
X_val        = MMS.fit_transform(x_val)

LB = LabelBinarizer()

Y_train_fair = LB.fit_transform(fair_y_train)
Y_val        = LB.fit_transform(y_val)
Y_test       = LB.fit_transform(y_test)
```

# The model

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, LeakyReLU
from keras.optimizers import Adam, SGD
from sklearn.metrics import accuracy_score
from tensorflow.keras.callbacks import EarlyStopping
from keras.layers.normalization import BatchNormalization


fair = Sequential()
#INPUT LAYER
fair.add(Dense(256, input_dim=X_train_fair.shape[1], activation='tanh'))


#HIDDEN LAYERS
fair.add(Dense(128)) 
fair.add(LeakyReLU(alpha=0.1))
fair.add(Dropout(0.1))

fair.add(Dense(64)) 
fair.add(LeakyReLU(alpha=0.1))
fair.add(Dropout(0.1))

fair.add(Dense(64)) 
fair.add(LeakyReLU(alpha=0.1))
fair.add(Dropout(0.1))

fair.add(Dense(32))
fair.add(LeakyReLU(alpha=0.1))


#OUTPUT LAYER
fair.add(Dense(1, activation='sigmoid'))


#run the model
optimizer = Adam(lr=0.0001) 
fair.compile(optimizer=optimizer, loss="binary_crossentropy", metrics="accuracy")
history = fair.fit(X_train_fair, Y_train_fair, epochs=7, batch_size=X_train_fair.shape[1], verbose=1, 
                       validation_data=(X_test,Y_test), callbacks=[EarlyStopping(monitor='val_loss', patience=5)])

```

```python
import matplotlib.pyplot as plt 

#accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

#loss 
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

# Balanced Residuals

```python
y_pred_val   = (fair.predict(X_val) > 0.5).astype("int32")
y_hats_val   = pd.DataFrame(data = y_pred_val, columns=['y_pred_val'], index=x_val.index.copy())
```

```python
regular_out = pd.merge(x_recid_new2, y_hats_val, how='left', left_index=True, right_index=True )

regular_out["y_val"] = y_val

display(regular_out)
```

```python
age1 = regular_out[regular_out["Age_Release"] == 1]
age2 = regular_out[regular_out["Age_Release"] == 2]

age_young = age1
age_old = age2

display(age_young)
display(age_old)
```

```python
age_old_ypredval = age_old["y_pred_val"]
age_old_ytrueval = age_old["y_val"]

age_young_ypredval = age_young["y_pred_val"]
age_young_ytrueval = age_young["y_val"]

age_old_ypredval = age_old_ypredval.dropna()
age_old_ytrueval = age_old_ytrueval.dropna()

age_young_ypredval = age_young_ypredval.dropna()
age_young_ytrueval = age_young_ytrueval.dropna()

display(age_old_ypredval)
display(age_old_ytrueval)
display(age_young_ypredval)
display(age_young_ytrueval)
```

```python
# TRUE - PRED
val_dif_old = (age_old_ytrueval - age_old_ypredval).sum()
print(val_dif_old)

residuals_old = val_dif_old / 2257
print(residuals_old)

# TRUE - PRED
val_dif_young = (age_young_ytrueval - age_young_ypredval).sum()
print(val_dif_young)

residuals_young = val_dif_young / 2190 
print(residuals_young)
```

```python
discrimination_balanced_residuals = residuals_young - residuals_old

print(discrimination_balanced_residuals)
```
