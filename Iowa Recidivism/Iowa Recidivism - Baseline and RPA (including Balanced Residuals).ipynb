# Transforming

```python
#train test split - import 
from sklearn.model_selection import train_test_split as tts

x, x_test, y, y_test = tts(x_recid, recid_target, test_size=0.20, train_size=0.80, random_state=33) 
x_train, x_val, y_train, y_val = tts(x, y, test_size=0.25, train_size=0.75, random_state=33)
```

```python
print("Train size:", x_train.shape)
print("Test size:", x_test.shape)
print("Validation size:", x_val.shape)
```

```python
from sklearn.preprocessing import LabelBinarizer, StandardScaler, LabelEncoder, MinMaxScaler

MMS = MinMaxScaler()

#SS = StandardScaler()
X_train = MMS.fit_transform(x_train)
X_test  = MMS.fit_transform(x_test)
X_val   = MMS.fit_transform(x_val)

LB = LabelBinarizer()
#LE = LabelEncoder()

Y_train = LB.fit_transform(y_train)
Y_val   = LB.fit_transform(y_val)
Y_test  = LB.fit_transform(y_test)
```

# The model

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, LeakyReLU
from keras.optimizers import Adam, SGD
from sklearn.metrics import accuracy_score
from tensorflow.keras.callbacks import EarlyStopping
from keras.layers.normalization import BatchNormalization


baseline = Sequential()
#INPUT LAYER
baseline.add(Dense(256, input_dim=X_train.shape[1], activation='tanh'))


#HIDDEN LAYERS
baseline.add(Dense(128)) 
baseline.add(LeakyReLU(alpha=0.1))
baseline.add(Dropout(0.1))

baseline.add(Dense(64)) 
baseline.add(LeakyReLU(alpha=0.1))
baseline.add(Dropout(0.1))

baseline.add(Dense(64)) 
baseline.add(LeakyReLU(alpha=0.1))
baseline.add(Dropout(0.1))

baseline.add(Dense(32))
baseline.add(LeakyReLU(alpha=0.1))


#OUTPUT LAYER
baseline.add(Dense(1, activation='sigmoid'))


#run the model
optimizer = Adam(lr=0.0001) 
baseline.compile(optimizer=optimizer, loss="binary_crossentropy", metrics="accuracy")
history = baseline.fit(X_train, Y_train, epochs=40, batch_size=X_train.shape[1], verbose=1, 
                       validation_data=(X_test,Y_test), callbacks=[EarlyStopping(monitor='val_loss', patience=5)])
```

```python
import matplotlib.pyplot as plt 

#accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

#loss 
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
#train test split - import 
from sklearn.model_selection import train_test_split as tts

x_pro, x_test_pro, y_pro, y_test_pro = tts(protected_recid, recid_target, test_size=0.20, train_size=0.80, random_state=33) 
x_train_pro, x_val_pro, y_train_pro, y_val_pro = tts(x_pro, y_pro, test_size=0.25, train_size=0.75, random_state=33)
```

```python
print("Train size:", x_train_pro.shape)
print("Test size:", x_test_pro.shape)
print("Validation size:", x_val_pro.shape)
```

```python
from sklearn.preprocessing import LabelBinarizer, StandardScaler, MinMaxScaler

#Without the StandardScaler on the dataset, the model is stuck in a local / global minima. 
#Standardizing didn't only solve the model being stuck, the performance also increased significantly after applying this.
MMS = MinMaxScaler()
X_train_pro = MMS.fit_transform(x_train_pro)
X_test_pro  = MMS.fit_transform(x_test_pro)
X_val_pro   = MMS.fit_transform(x_val_pro)

LB = LabelBinarizer()

Y_train_pro = LB.fit_transform(y_train_pro)
Y_val_pro   = LB.fit_transform(y_val_pro)
Y_test_pro  = LB.fit_transform(y_test_pro)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, LeakyReLU
from keras.optimizers import Adam, SGD
from sklearn.metrics import accuracy_score
from tensorflow.keras.callbacks import EarlyStopping
from keras.layers.normalization import BatchNormalization


protected = Sequential()
#INPUT LAYER
protected.add(Dense(256, input_dim=X_train_pro.shape[1], activation='tanh'))


#HIDDEN LAYERS
protected.add(Dense(128)) 
protected.add(LeakyReLU(alpha=0.1))
protected.add(Dropout(0.1))

protected.add(Dense(64)) 
protected.add(LeakyReLU(alpha=0.1))
protected.add(Dropout(0.1))

protected.add(Dense(64)) 
protected.add(LeakyReLU(alpha=0.1))
protected.add(Dropout(0.1))

protected.add(Dense(32))
protected.add(LeakyReLU(alpha=0.1))


#OUTPUT LAYER
protected.add(Dense(1, activation='sigmoid'))


#run the model
optimizer = Adam(lr=0.0001) 
protected.compile(optimizer=optimizer, loss="binary_crossentropy", metrics="accuracy")
history_protected = protected.fit(X_train_pro, Y_train_pro, epochs=40, batch_size=X_train_pro.shape[1], verbose=1, 
                       validation_data=(X_test_pro,Y_test_pro), callbacks=[EarlyStopping(monitor='val_loss', patience=5)])
```

```python
#accuracy
plt.plot(history_protected.history['accuracy'])
plt.plot(history_protected.history['val_accuracy'])
plt.title('Protected model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

#loss 
plt.plot(history_protected.history['loss'])
plt.plot(history_protected.history['val_loss'])
plt.title('Protected model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

# Balanced Residuals

## baseline

```python
y_pred_val   = (baseline.predict(X_val) > 0.5).astype("int32")
y_hats_val   = pd.DataFrame(data = y_pred_val, columns=['y_pred_val'], index=x_val.index.copy())
```

```python
regular_out = pd.merge(x_recid, y_hats_val, how='left', left_index=True, right_index=True )

regular_out["y_val"] = y_val

display(regular_out)
```

```python
cleanup_nums1 = {"y_pred_val": {0.0:1, 1.0:2},
                 "y_val": {0.0:1, 1.0:2}}

regular_out = regular_out.replace(cleanup_nums1)
display(regular_out)
```

```python
age1 = regular_out[regular_out["Age_Release"] == 1]
age2 = regular_out[regular_out["Age_Release"] == 2]

age_young = pd.concat([age1, age2])
display(age_young)

age3 = regular_out[regular_out["Age_Release"] == 3]
age4 = regular_out[regular_out["Age_Release"] == 4]
age5 = regular_out[regular_out["Age_Release"] == 5]

age_old = pd.concat([age3, age4, age5])
display(age_old)
```

```python
age_old_ypredval = age_old["y_pred_val"]
age_old_ytrueval = age_old["y_val"]

age_young_ypredval = age_young["y_pred_val"]
age_young_ytrueval = age_young["y_val"]

age_old_ypredval = age_old_ypredval.dropna()
age_old_ytrueval = age_old_ytrueval.dropna()

age_young_ypredval = age_young_ypredval.dropna()
age_young_ytrueval = age_young_ytrueval.dropna()

display(age_old_ypredval)
display(age_old_ytrueval)
display(age_young_ypredval)
display(age_young_ytrueval)
```

```python
# TRUE - PRED
val_dif_old = (age_old_ytrueval - age_old_ypredval).sum()
print(val_dif_old)

residuals_old = val_dif_old / 2227
print(residuals_old)

# TRUE - PRED
val_dif_young = (age_young_ytrueval - age_young_ypredval).sum()
print(val_dif_young)

residuals_young = val_dif_young / 2618 
print(residuals_young)
```

```python
discrimination_balanced_residuals = residuals_young - residuals_old

print(discrimination_balanced_residuals)
```
